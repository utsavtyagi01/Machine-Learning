```python
import pandas as pd
```

```python
df=pd.read_csv(r'c:\Users\Utsav\Downloads\titanic.csv')
```

```python
df.drop(['PassengerId','Embarked','Cabin','Name','SibSp','Parch','Ticket'],axis="columns",inplace=True)
```

```python
df
```

```python
input=df.drop(["Survived"],axis="columns")
target=df.Survived
```

```python
dummy=pd.get_dummies(input.Sex).astype(int)
input=pd.concat([input,dummy],axis="columns")
input.drop(['Sex'],axis="columns",inplace=True)
```

```python
input
```

```python
input.columns[input.isna().any()]  #This executes the function in square brackets corresponds to every column.
```

```python
mean=input.Age.mean()
input.Age.fillna(mean,inplace=True)
```

```python
input
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict_log_proba(x_test)
```
