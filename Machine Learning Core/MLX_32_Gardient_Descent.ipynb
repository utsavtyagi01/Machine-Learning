```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y=make_regression(n_samples=4,n_informative=1,n_features=1,n_targets=1,noise=80,random_state=13)
```

```python
x.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='r')
```

Applying Gradient Descent assuming slope is constant= 78.35 <br>
and b=0

```python
y_pred=((78.35*x)+0).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x),color='r',label='OLS')
plt.plot(x,y_pred,color='g',label='b=0')
plt.legend()
plt.show()
```

```python
m=78.35
b=0
```

```python
slope = -2 *np.sum(y-m*x.ravel()-b) #ravel is used to flatten multidimensional array
```

```python
lr=0.1

step_size=slope*lr
step_size
```

```python
# calculating new intercept

b= b-step_size
b
```

```python
#Agin plotting graph with new b

y_pred1=((78.35*x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,model.predict(x),color='black',label='OLS')
plt.plot(x,y_pred1,color='pink',label='b = {}'.format(b))
plt.plot(x,y_pred,color='r',label='b=0')
plt.legend()
plt.show()
```

```python
step_size=slope*lr
step_size
```

```python
b= b-step_size
b
```

```python
y_pred2=((78.35*x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,model.predict(x),color='black',label='OLS')
plt.plot(x,y_pred2,color='yellow',label='b = {}'.format(b))
plt.plot(x,y_pred1,color='pink',label='b = {}'.format(b))
plt.plot(x,y_pred,color='r',label='b=0')
plt.legend()
plt.show()
```

```python
step_size=slope*lr
step_size
```

```python
b= b-step_size
b
```

```python
y_pred3=((78.35*x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,model.predict(x),color='black',label='OLS')
plt.plot(x,y_pred3,color='green',label='b = {}'.format(b))
plt.plot(x,y_pred2,color='yellow',label='b = {}'.format(b))
plt.plot(x,y_pred1,color='pink',label='b = {}'.format(b))
plt.plot(x,y_pred,color='r',label='b=0')
plt.legend()
plt.show()
```

```python
m=78.35
b=-100
lr=0.01
epochs=100
for i in range(epochs):
    slope = -2 *np.sum(y-m*x.ravel()-b)
    step_size=slope*lr
    b= b-step_size
    y_pred=m*x+b
    plt.plot(x,y_pred)
plt.scatter(x,y)
```
