Encoding with greater number of categories

```python
count=df.brand.value_counts() #yahan pe humne saari cars brand ki values ko store kr liya hain
```

```python
replace=count[count<=100] 
#yahan pe fir jinki value 100 se kam h unko alag variable mein store kr liya hain
```

```python
df.brand=df.brand.apply(lambda x: 'others' if x in replace else x)
#yahan pe humne jo brand replace mein the unko others se replace krdia h taaki categories kam ho jaein
```
