```python
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

1. extract year

```python
date['year']=date['date'].dt.year
date.sample(5)
```

2.extract month

```python
date['month']=date['date'].dt.month
date.sample(5)
```

```python
date['month']=date['date'].dt.month_name()
date.sample(5)
```

3. extract day of week

```python
date['day']=date['date'].dt.dayofweek
date.sample(5)
```

```python
date['day_name']=date['date'].dt.day_name()
date.sample(5)
```

```python
# IS WEEKEND

date['Is_Weekend']=np.where(date['day_name'].isin(['Sunday','Saturday']),1,0)
```

```python
date
```

Extract Week of the year

```python
date['week']=date['date'].dt.isocalendar().week
date
```

Extract time elapsed between dates

```python
import datetime

today = datetime.datetime.today()
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

np.round((today -date['date']) / np.timedelta64(1, 'M'),0)
```
