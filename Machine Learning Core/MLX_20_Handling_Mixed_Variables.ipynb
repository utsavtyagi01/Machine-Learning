```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('titanic1.csv')
```

```python
df
```

```python
df.number.unique()
```

```python
# Extract numerical part
df['numerical']=pd.to_numeric(df.number,errors='coerce',downcast='integer')
```

```python
df
```

```python
# extract categorical data
df['categorical'] = np.where(df.numerical.isnull(),df.number,np.nan)
```

```python
df
```

```python
df['cabin_num']=df['Cabin'].str.extract('(\d+)') #captures numerical part
df['cabin_cat']=df['Cabin'].str[0] #captures first letter
```

```python
df
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat']) #jahan pe digit hai wahan pe na krdo 
```
