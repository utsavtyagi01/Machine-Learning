```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv('titanic.csv')[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.head()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df.dropna(inplace=True)
```

```python
np.mean(cross_val_score(LogisticRegression(),df.iloc[:,0:4],df.iloc[:,4],scoring='accuracy',cv=20))
```

Applying Feature Construction

```python
df['Family_Size']=df['SibSp']+df['Parch']+1
```

```python
df
```

```python
def func(temp):
    if(temp==1):
        return 0
    elif(temp>1 and temp<=4):
        return 1
    else:
        return 2
```

```python
df['Family_Type']=df['Family_Size'].apply(func)
```

```python
df
```

```python
df.drop(['SibSp',"Parch",'Family_Size'],axis=1,inplace=True)
```

```python
df
```

```python
np.mean(cross_val_score(LogisticRegression(),df.iloc[:,[0,1,3]],df.iloc[:,2],scoring='accuracy',cv=20))
```
