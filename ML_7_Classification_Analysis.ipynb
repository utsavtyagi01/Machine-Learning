```python
df.drop(['Department','salary'],axis="columns").groupby('left').mean()
```

```python
import seaborn as sns
```

```python
count_df = df.groupby(['salary', 'left']).size().reset_index(name='count')

sns.set(style="whitegrid")
g = sns.catplot(x='salary', y='count', hue='left', data=count_df, kind='bar')

g.set_axis_labels('Salary', 'Number of People')
g.set(title='Number of People Left vs. Not Left by Salary')

# Show the plot
plt.show()
```
