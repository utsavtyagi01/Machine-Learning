```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df=pd.read_csv('house-prices.csv')
```

```python
df1=df.drop(columns=['Home','Bedrooms','Offers','Brick','Neighborhood','Bathrooms'])
```

```python
df.head()
```

```python
plt.scatter(df.SqFt,df.Price)
plt.show()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(data=df,x='SqFt',y='Price')
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
```

```python
df1
```

```python
model.fit(df1[['SqFt']],df1.Price) #here we will give a 2d array to X as Linear Regression also works on multiple features so in order to be compatible be with all course of data it
                                    # does not uses 1d array
```

```python
import numpy as np
y_pred=model.predict(np.array(df1['SqFt']).reshape(-1, 1)) # here also it takes 2d array as input
```

```python
plt.scatter(df1.SqFt,df1.Price)
plt.plot(df1.SqFt,y_pred,color='red')
plt.show()                            #np.reshape is not inplace
```

```python
# Using Test Train Split
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(df1.SqFt,df1.Price,test_size=0.2)
```

```python
model.fit(np.array(x_train).reshape(-1,1),y_train)
```

```python
y_pred=model.predict(np.array(y_test).reshape(-1,1))
```

```python
y_test
```

```python
plt.scatter(x_train, y_train, label='Training Set', alpha=0.7)
plt.scatter(x_test, y_test, label='Testing Set', alpha=0.7)
plt.title('Train-Test Split')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
plt.scatter(x_train, y_train, label='Training Set', alpha=0.7)
plt.scatter(x_test, y_test, label='Testing Set', alpha=0.7)
plt.plot(x_train, model.predict(np.array(x_train).reshape(-1, 1)), color='red', label='Regression Line')
plt.title('Train-Test Split with Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
